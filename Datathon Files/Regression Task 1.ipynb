```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
descriptors = pd.read_csv('train_descriptors.csv')
descriptors
```

```python
train_crystals=np.loadtxt('train_crystals2.csv',delimiter=',', skiprows=1)
print(train_crystals)
```

```python
y = pd.read_csv('train_crystals2.csv')
```

```python
descriptor_list = []
for descriptor in descriptors.columns:
    if descriptor != 'identifiers' and descriptor != 'SMILES' and descriptor != 'name':
        try:
            corr = descriptors[descriptor].corr(y['calculated_density'])
            if abs(corr) > 0.7:
                print(f"{descriptor}: {corr}")
                descriptor_list.append(descriptor)
        except:
            pass
del descriptor_list[14]
del descriptor_list[16]
```

```python
l = []
for d in descriptor_list:
    l.append(descriptors[d].to_numpy())
```

```python
density = train_crystals[:,9]
```

```python
def normalise(x):
    mean = np.nanmean(x)
    x = np.nan_to_num(x, mean)
    x_max = np.amax(x)
    x_min = np.amin(x)
    x_n = (x-x_min)/(x_max-x_min)
    return x_n
```

```python
for i in range(0, len(l)):
    l[i] = normalise(l[i])
```

```python
density_n = normalise(density)
density_n = density_n.reshape(-1, 1)
```

```python
X = np.array([])
for m in range(0,len(l[0])):
    new_array = np.array([])
    for j in range(0,len(descriptor_list)):
        new_array = np.append(new_array,l[j][m])
    X = np.append(X,new_array)
print(X)
```

```python
X = np.reshape(X,(13449,len(descriptor_list)))
```

```python
m = int(0.8*13449)
X_train = X[:m,:]
X_test = X[m:,:]
Y_train = density_n[:m,:]
Y_test = density_n[m:,:]
```

```python
X_train
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train,Y_train)
```

```python
y_pred = linear_regression.predict(X_test)
```

```python
mean_absolute_error(Y_test, y_pred)
```

```python
print(y_pred)
```

```python
np.savetxt('task_1_test.csv', y_pred, delimiter=',')
```
